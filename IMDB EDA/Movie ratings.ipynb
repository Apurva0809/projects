<b>Which are the movies with the third lowest and third highest budget?</b>

```python
df_1 = df[df.budget > 0]
df_1
```

<b>Top 5 budgets are:</b>

```python
df_1_1 = df_1.sort_values('budget', ascending=False).head()
df_1_1[['original_title', 'budget', 'revenue']].reset_index()
```

```python
x = df_1_1['original_title']
y_pos = np.arange(len(x))
y = df_1_1['budget']

plt.figure(figsize=(20,9))
plt.bar(x, y, align='center', alpha=0.5)
plt.xticks(y_pos, x, fontsize=10)
plt.ylabel('Budgets', fontsize=12)
plt.title('Movies with highest Budgets', fontsize=20)
 
plt.show()
```

<b>Least budget movies are:</b>

```python
df_1_2 = df_1.sort_values('budget').head(10)
df_1_2[['original_title', 'budget','revenue']].reset_index()
```

```python
x_1 = df_1_2['original_title']
y_pos_1 = np.arange(len(x_1))
y_1 = df_1_2['budget']

plt.figure(figsize=(30,10))
plt.bar(x_1, y_1, align='center', alpha=0.5)
plt.xticks(y_pos_1, x_1, fontsize=12)
plt.ylabel('Budgets', fontsize=12)
plt.title('Movies with lowest Budgets',fontsize=20)
 
plt.show()
```

<b>Which are the movies with most and least earned revenue?</b>

```python
df_4 = df[df.revenue > 0]
df_4
```

1. Movies with most earned revenues:

```python
df_4_1=df_4.sort_values('revenue', ascending = False).head()
df_4_1[['original_title', 'budget', 'revenue']]
```

```python
x_2 = df_4_1['original_title']
y_pos_2 = np.arange(len(x_2))
y_2 = df_4_1['revenue']

plt.figure(figsize=(15,8))
plt.bar(x_2, y_2, align='center', alpha=0.5)
plt.xticks(y_pos_2, x_2, fontsize=12)
plt.ylabel('Revenues', fontsize=12)
plt.title('Movies with highest Revenues', fontsize=20)
 
plt.show()
```

2. Movies with least earned revenues:

```python
df_4_2=df_4.sort_values('revenue').head()
df_4_2[['original_title','budget','revenue']]
```

```python
x_3 = df_4_2['original_title']
y_pos_3 = np.arange(len(x_3))
y_3 = df_4_2['revenue']

plt.figure(figsize=(15,9))
plt.bar(x_3, y_3, align='center', alpha=0.5)
plt.xticks(y_pos_3, x_3, fontsize=12)
plt.ylabel('Revenues', fontsize=12)
plt.title('Movies with lowest earned revenues', fontsize=20)
 
plt.show()
```

<b>Production companies which have invested money in worse revenue movies.</b>

```python
df_6 =df[df.revenue > 0]
df_6
```

```python
df_6_1 = df_6.sort_values('revenue').head(10)
df_6_2 = df_6_1[['revenue','original_title','production_companies']]
df_6_2.dropna(subset=['production_companies'],how='any',inplace=True)
```

```python
x_4 = df_6_2['production_companies']
y_pos_4 = np.arange(len(x_4))
y_4 = df_6_2['revenue']

plt.figure(figsize=(20,9))
plt.bar(x_4, y_4, align='center', alpha=0.5)
plt.xticks(y_pos_4, x_4, fontsize=8)
plt.ylabel('Revenues', fontsize=12)
plt.title('Production companies which invested in the worst revenue movies', fontsize=20)
 
plt.show()
```
